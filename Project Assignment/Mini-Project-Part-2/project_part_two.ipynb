```python
# Dependencies
import pandas as pd
```

```python
# load CSV
file = 'Resources/2016-FCC-New-Coders-Survey-Data.csv'
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
df_raw = pd.read_csv(file)
df_raw.head()
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
cols = df_raw.columns[[0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
df1 = df_raw.loc[:, cols]
df1.head()
```

```python
# Change reading 0 and 1 to No and Yes, respectively
df2 = df1.replace({0:'No', 1:'Yes'})
```

```python
# Extract rows for only those who attended a bootcamp
filt1 = df2['AttendedBootcamp'] == "Yes"
bootcamp_df = df2.loc[filt1]
bootcamp_df.head()
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
bootcamp_group1 = bootcamp_df.groupby(['BootcampName'])
num_of_attendees = bootcamp_group1['AttendedBootcamp'].count()

attendees_df = pd.DataFrame(num_of_attendees)
attendees_df.head()
```

```python
# binning and create new data frame
bins=[0,10,100]
group_labels= ['<10 Attendees', 'Significant']

pd.cut(attendees_df['AttendedBootcamp'], bins, labels=group_labels)
attendees_df['Number of Attendees'] = pd.cut(attendees_df['AttendedBootcamp'], bins, labels=group_labels)
attendees_df.head()
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
recommend = bootcamp_group1['BootcampRecommend'].count()
recommend_df = pd.DataFrame(recommend)
recommend_df.head()
```

```python
# Merge the two created data frames on the name of tbe bootcamp
merge_df = pd.merge(attendees_df, recommend_df, on='BootcampName')
merge_df.head()
```

```python
# Calculate percentage of eac bootcamp's students who are recommenders
percentage = (merge_df['BootcampRecommend'] / merge_df['AttendedBootcamp'])*100
percentage_df = pd.DataFrame(percentage)

# merge three data frames
merge_df2 = pd.merge(merge_df, percentage_df, on='BootcampName')
renamed_merge = merge_df2.rename(columns={0:"Recommended Percentage"})

bootcamp_final_df = renamed_merge.drop(index='Free Code Camp is not a bootcamp - please scroll up and change answer to "no"')
bootcamp_final_df.head()
```

```python
# Export to excel and remove index
bootcamp_final_df.to_csv('Resources/Updated_2016_Bootcamp_Eval.csv', index=False, header=True)
```
