```python
# Import Dependencies
import pandas as pd
```

```python
# File to Load
data_one = "Resources/books_clean.csv"
# Read the modified GoodReads csv and store into Pandas DataFrame
data_one_df = pd.read_csv(data_one, encoding="ISO-8859-1")
data_one_df.head()
```

```python
# Calculate the number of unique authors in the DataFrame
authors = data_one_df["Authors"].unique()
authors
```

```python
# Calculate the earliest/latest year a book was published
earliest_year = data_one_df['Publication Year'].min()
earliest_year

latest_year = data_one_df['Publication Year'].max()
latest_year
```

```python
# Calculate the total reviews for the entire dataset
# Hint: use the pandas' sum() method to get the sum for each row
total_reviews = data_one_df.iloc[:, 4:].sum(axis=1)
data_one_df["Total Review"] = total_reviews
```

```python
# Place all of the data found into a summary DataFrame
data_one_df.to_csv("Resources/books_reviewed.csv", index=False, header=True)
```
